```python
#in case if null value present in more columns, we can use imputer
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtypes=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual= quanQual(dataset)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
dataset_quan=imp.transform(dataset[quan])
#using imputer function for filling values in more than one column with central tendency
```

```python
desc=pd.DataFrame(dataset_quan,columns=quan)
```

```python
desc
#converted into table
#nan values filled by central tendency method(mean) using imputer
```

```python
df=dataset[qual]
```

```python
two=[desc,df]
```

```python
preprocessed=pd.concat(two,axis=1)#axis=1 is to print column wise
```
